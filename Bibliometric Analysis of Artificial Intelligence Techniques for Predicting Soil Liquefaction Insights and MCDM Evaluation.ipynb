# Bibliometric Analysis of Artificial Intelligence Techniques for Predicting Soil Liquefaction: Insights and MCDM Evaluation

## Introduction

This notebook contains the computational code associated with the research article titled "Bibliometric Analysis of Artificial Intelligence Techniques for Predicting Soil Liquefaction: Insights and MCDM Evaluation". The study provides a comprehensive bibliometric analysis of the literature on the application of AI techniques in predicting soil liquefaction. A systematic review was conducted on articles sourced from the Web of Science database spanning from 1994 to 2023.

## Abstract

The geotechnical phenomenon of soil liquefaction poses significant risks to infrastructure and human safety, necessitating effective prediction and mitigation strategies. With urbanization and infrastructural development expanding, there has been an increased focus on leveraging artificial intelligence (AI) techniques to address these complex geotechnical challenges. This notebook details the application of bibliometric indicators and the MULTIMOORA method, a Multi-Criteria Decision Making (MCDM) technique, to analyze trends, authorship patterns, affiliated institutions, publication venues, and citation patterns among the collected articles.

The analysis underscores the interdisciplinary nature of this field, merging geotechnical engineering, computer science, and machine learning. It also highlights significant journals like the Soil Dynamics and Earthquake Engineering journal, among others. A noticeable increase in publications was observed in 2011 and 2019, indicating growing interest in the application of AI in soil liquefaction studies.
